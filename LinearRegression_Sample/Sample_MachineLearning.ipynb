```python
d = {'col1': [1, 2, 3,4,5], 'col2': [3, 4,6,7,8]}
df = pd.DataFrame(data=d)
df
```

```python
def SumOfErrors(Y_actual, Y_predict):
    N = float(len(Y_actual))
    errors = (Y_actual - Y_predict)
    sumOfErrors = 0
    for i in errors:
        sumOfErrors += i**2
    return sumOfErrors/ N
```

```python
def GetWeightsAndBias(X,Y_actual, Y_predict):
    N = float(len(X))
    error = (Y_actual - Y_predict)
    m_gradient = - (2/N)* sum(X*error)
    b_gradient = -(2/N)* sum(error)
    return m_gradient,b_gradient
```
